```python
x=int(input('enter amount'))#to ask for user's bank balance
y=int(input("enter pin"))# to ask for user's pin
pin=1234#default pin
balance=50000 #default balance
print("===== Welcome to ATM Kiosk =====")
if y == pin:#condition to check the pin
    print("Login Successful\n")#if pin is matching the default pin then it will display this

    while True:
        print("\nSelect Transaction")
        print("1. Check Balance")
        print("2. Withdraw Money")
        print("3. Deposit Money")
        print("4. Exit")

        choice = input("Enter your choice: ")# to take choice from user

        if choice == "1":
            print("Your Balance is:", balance)#to check balance(current)

        elif choice == "2":#to enter amount to be withdrawn
            amount = int(input("Enter amount to withdraw: "))
            if amount <= balance:#only possibility to withdraw amount
                balance -= amount#it will deduct the amt entered
                print("Transaction Successful")
                print("Remaining Balance:", balance)#current bank balance-entered amt
            else:
                print("Insufficient Balance")#if entered amt is more than the current amt it will show this

        elif choice == "3":#to enter amount to be deposited
            amount = int(input("Enter amount to deposit: "))
            balance += amount
            print("Deposit Successful")
            print("New Balance:", balance)

        elif choice == "4":#to exit choices
            print("Thank you for using ATM")
            break

        else:
            print("Invalid Choice")

else:
    print("Incorrect PIN")
```
